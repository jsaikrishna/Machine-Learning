```python
def removepunc1(text):
    punctuations = '''!()-[]{};:'"\,<>./?@#$%^&*_~'''
    for x1 in text:
        no_punct = ""
        for char in x1:
            if char not in punctuations:
                no_punct = no_punct + char
        x1=no_punct
    return text
```

```python
import nltk
from nltk import word_tokenize
from nltk.stem import WordNetLemmatizer, PorterStemmer ,SnowballStemmer

class LemmaTokenizer(object):
    def __init__(self):
        self.wnl = WordNetLemmatizer()
        self.port= PorterStemmer()
        self.sno=  SnowballStemmer('english')
    def __call__(self, ex):
        #stem= [self.port.stem(t) for t in word_tokenize(doc)]
        #lem=[self.wnl.lemmatize(t) for t in word_tokenize(doc)]
        #snow=[self.sno.stem(t) for t in word_tokenize(doc)]
        
        return [self.port.stem(t) for t in word_tokenize(ex)]
    
```

```python
from sklearn.base import BaseEstimator, TransformerMixin
class capTransformer(BaseEstimator, TransformerMixin):
    
    def __init__(self):
        pass
    
    def fit(self, examples):
        # return self and nothing else 
        return self
    
    def get_feature_names(self):
        self.feature_names = ["Capital"]
        return self.feature_names
    
    def transform(self, examples):
        
        import numpy as np 
        from scipy.sparse import csr_matrix
        
        Z1 = np.zeros((len(examples), 1))
        
        # Loop over examples and count letters
        for i,st in enumerate(examples):
            Z1[i,:]=np.array((len(re.findall(r'[A-Z]',st))))
            #print(np.array([wx[key] for key in wx]))
            
        return csr_matrix(Z1) 
```

```python
from sklearn.base import BaseEstimator, TransformerMixin

class lengthTransformer(BaseEstimator, TransformerMixin):
    
    def __init__(self):
        pass
    
    def fit(self, examples):
        # return self and nothing else 
        return self
    
    def get_feature_names(self):
        self.feature_names = ["Length Of Sentence"]
        return self.feature_names
        
    def transform(self, examples):
        
        import numpy as np 
        from scipy.sparse import csr_matrix
        
         
        Y1 = np.zeros((len(examples), 1))
        
        # Loop over examples and count letters 
        for j, y1 in enumerate(examples):
            Y1[j,:] = np.array(len(y1))
        #print("Hi")
            
        return csr_matrix(Y1) 
```

```python
from sklearn.base import BaseEstimator, TransformerMixin
class VocabularyCounter(BaseEstimator, TransformerMixin):
    def __init__(self):
        pass
    def fit(self, examples):
        return self
    
    def get_feature_names(self):
        self.feature_names = ["Vocabulary count"]
        return self.feature_names

    def transform(self, examples):
        
        import numpy as np 
        from scipy.sparse import csr_matrix
        
        f = [dict() for x in range(len(examples))]
        i = 0
        for ex in examples:
            wv = re.compile('\b\w+\w').split(str(ex))
            for w in wv:
                if w in f[i].keys():
                    f[i][w] += 1
                else:
                    f[i][w] = 1
            i += 1
        v = np.zeros((len(examples), 1))
        i = 0
        for f1 in f:
            v[i] = len(f1.keys())
            i += 1
            
        return csr_matrix(v)
```
